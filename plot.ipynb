```python
import epa_graphs as eg
```

```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as pp
import collections
```

```python
pp.rcParams['figure.figsize'] = 14,7
```

```python
graph = eg.gen_directed_graph(4)
G = nx.DiGraph()
edge_labels = collections.defaultdict(str)
for node_1 in graph:
    for node_2,cost in graph[node_1]:
        G.add_edges_from([(node_1,node_2)])
        weight = sorted([node_1,node_2])
        edge_labels[tuple(weight)] += "{:.2f},".format(cost)

pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G,pos)
nx.draw_networkx_edges(G,pos)
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels);
```

```python
graph = eg.gen_undirected_graph(10,9)
G = nx.DiGraph()
edge_labels = collections.defaultdict(str)
for node_1 in graph:
    for node_2,cost in graph[node_1]:
        G.add_edges_from([(node_1,node_2)])
        weight = sorted([node_1,node_2])
        edge_labels[tuple(weight)] = "{:.2f}".format(cost)

pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G,pos)
nx.draw_networkx_edges(G,pos)
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels);
```
